### 1. Porównanie błędów względnych algorytmów dla typu Float64 przy k=54 na brzegach dziedziny funkcji

```julia
epsilons()
```

Zaprezentowany powyżej algorytm liczy maksymalne błędy względne(maksimum z błędu względnego dla części rzeczywistej i urojonej) dla odpowiednio: małych, średnich i dużych wartości oraz k:=54(wybór Briggsa). Dla każdej pary zakresów (ε) generuje równo oddalone próbki z zakresu [-ε, +ε] zarówno dla części rzeczywistej jak i urojonej. Następnie zlicza największe względne błędy przybliżeń i zapisuje je w tablicy relErrors. Możemy zauważyć, że rzędy błędów dla algorytmu 1(oryginalnego) w porównaniu do algorytmu 2(Al-Mohy) przy k=54 dla typu Float64(podwójna precyzja) są o wiele większe dla każdych par zakresów. Można przetestować wynik dla różnych wielkości k podając do funkcji `epsilons()` parametr `maxk::Integer`.

### 2. Porównanie błędów względnych algorytmów dla typu Float64 oraz rosnących k na liczbach zespolonych.

```julia
epsilonK(maxk=70)
```

Powyższy algorytm liczy wartość błędu względnego dla algorytmów 1 i 2 dla typu Float64 na zmiennych zespolonych w otoczeniu ε=10 od punktu 0+0i przy 200 próbkach i 50 iteracjach. Jak widać na załączonym wykresie, już po 26 iteracjach algorytm 1 zaczyna tracić na dokładności nie potrafiąc obliczyć wyników z dokładnością większą niż 8 cyfr dziesiętnych. Można przetestować działanie `epsilonK()` dla innego otoczenia (parametr `ε`), innej maksymalnej liczby iteracji (parametr `maxk::Integer`), oraz innej ilości próbek (parametr `samples::Integer`). 

### 3. Porównanie błędów względnych algorytmów dla różnych k i typu BigFloat na liczbach rzeczywistych

Porównujemy błąd względny dla k=1:200 na liczbach rzeczywistych małych, średnich i dużych zakresów. 

```julia
increasingK()
```

Widać przewagę udoskonalonego algorytmu nad oryginalną metodą Briggsa: przy k większych niż 130 pierwsza metoda jest obarczona tak dużym błędem numerycznym, że każdka kolejna iteracja powoduje pogorszenie wyniku. Z tego powodu nie jesteśmy w stanie uzyskać przybliżeń dokładniejszych niż ≈10e-39 na liczbach typu BigFloat(256 bitów). Błąd drugiej metoda zbiega w tempie logarytmicznym do zera(liniowa zależność ilości dokładnych cyfr znaczących od k), dzięki czemu możemy uzyskiwać coraz lepsze przybliżenia. Warto może zastanowić się, jaką  dokładność możemy otrzymać samym algorytmem 2 i jak zachowa się on dla jeszcze większych k na liczbach rzeczywistych. Aby wypisywać maksymalne wartości błędów względnych można użyć parametru `logging=true`.
